```python
df1 = data1.to_frame(name="SP500")
df2 = data2.to_frame(name='RRPONTSYD')
```

```python
import pandas as pd
df_merged = pd.concat([df1, df2], axis=1)
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot(df_merged.index, df_merged['SP500'])
plt.plot(df_merged.index, df_merged['RRPONTSYD'])
```

```python
fig,ax = plt.subplots()
ax.plot(df_merged.index,
df_merged['SP500'],color="blue",marker="o")
ax2=ax.twinx()
ax2.plot(df_merged.index,
df_merged['RRPONTSYD'],color="red")
fig.savefig('.jpg',format='jpeg')
```
